```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0, y0, x1, y1 = np.loadtxt('classasgntrain1.dat').T
```

```python
fig, ax = plt.subplots()
ax.plot(x0, y0, 'gx', label='Class 0')
ax.plot(x1, y1, 'ro', label='Class 1')
ax.legend()
plt.title("Sample Population")
plt.show()
```
